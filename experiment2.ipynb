```python
# list items
# list items are ordered, changeable, allow duplicate values
# list items are indexed, the first item has index [0], the second item has index [1], and so on
# duplicates are allowed
names=["ananya","atharv","anshoo","atharv"]
print(names)
```

```python
# list length = no. of items in the list
names=["ananya","atharv","anshoo"]
print(len(names))
```

```python
# list can include items of different datatypes
list1=["abc",12,False,4389,"male",23.45,3+2j]
print(list1)
```

```python
# access items
marks=[20,14,18.5,6]
print(marks[3])
```

```python
# reverse indexing
marks=[20,14,18.5,6]
print(marks[-2])
```

```python
# slicing
# first index - inclusive; second index - exclusive
names=["ananya","atharv","anshoo","manisha","satya","prabha"]
print(names[2:5])
```

```python
# one index given
print(names[:4])
print(names[3:])
```

```python
# negative indexing in slicing
print(names[-4:-1])
print(names[2:-2])
print(names[-5:4])
print(names[-1:])
```

```python
# change item value
marks=[20,14,18.5,6,11]
marks[1]=13
print(marks)
marks[0]=19
print(marks)
```

```python
# append items
marks=[19, 13, 18.5, 6, 11]
marks.append(15)
print(marks)
```

```python
# insert items
marks=[19, 13, 18.5, 6, 11, 15]
marks.insert(2,7)
print(marks)
```

```python
# extend list
fruits=["apple","banana","cherry"]
fruits1=["orange","kiwi","papaya"]
fruits.extend(fruits1)
print(fruits)
```

```python
# functions on lists containing integers
# PRACTICE QUESTION 1
numbers=[10, 20, 30, 40, 50]

"minimum marks, maximum marks, total marks, no. of students, average marks"
print("minimum marks: ", min(numbers))
print("maximum marks: ", max(numbers))
print("total marks: ", sum(numbers))
print("no. of students: ", len(numbers))
print("average marks: ", (sum(numbers))/(len(numbers)))

"sort the marks in ascending order"
numbers.sort()
numbers
```

```python
# PRACTICE QUESTION 3
'''A person stores phone contacts in a list.
Tasks:
> Create a list of contacts
> Add a new contact
> Delete an existing contact
> Display contacts alphabetically'''

contactlist=["atharv","anshoo","manisha","satya","prabha"]
newcontacts=['nerissa','anusha','amog']
contactlist.extend(newcontacts)
print(contactlist)
contactlist.remove("amog")
print(contactlist)
contactlist.sort()
contactlist
```

```python
# PRACTICE QUESTION 4
'''Temperature Analysis
Daily temperature readings of a city are stored in a list for 10 days.
Tasks:
> Display temperatures of first and last 5 days
> Find highest and lowest temperature
> Calculate average temperature'''

temp=[35,26,7,40,34.2,42,28,34,46,12.1]
# temperature in first 5 days
print(temp[:5])
# temperature in last 5 days
print(temp[-5:])
# highest temperature
print(max(temp))
# lowest temperature
print(min(temp))
# average temperature
print((sum(temp)/len(temp)))
```

```python
#List Slicing
list1=[10,20,30,40,50,60]
print(list1[1:3])
print(list1[1:3:1])
print(list1[1:3:-1])
print(list1[2:5:2])
print(list1[1:4:3])
print(list1[-1:3:1])
print(list1[-1:3:-1])
```
